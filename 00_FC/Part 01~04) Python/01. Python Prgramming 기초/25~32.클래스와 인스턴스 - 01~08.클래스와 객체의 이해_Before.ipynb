#### **self**
- 파이썬의 method는 항상 첫번째 인자로 self를 전달
- self는 현재 해당 메쏘드가 호출되는 객체 자신을 가리킴
- C++/C#, JAVA의 this에 해당
- 역시, 이름이 self일 필요는 없으나, 위치는 항상 맨 처음의 parameter이며 관례적으로 self로 사용

```python
class Person: 
    def __init__(self, name, age):
        print('self:', self)
        self.name = name
        self.age = age
        
    def sleep(self):
        print('self', self)
        print(self.name, '은 잠을 잡니다.')
        
a = Person('Aaron', 20)
b = Person('Bob', 20)

print(a)
print(b)

a.sleep()
b.sleep()
```

#### **method type**
 - instance method - 객체로 호출
   - 메쏘드는 객체 레벨로 호출 되기 때문에, 해당 메쏘드를 호출한 객체에만 영향을 미침
 - class method    - class로 호출
       - 클래스 메쏘드의 경우, 클래스 레벨로 호출되기 때문에, 클래스 멤버 변수만 변경 가능


```python
class Math:
    def add(self, a, b):
        return a + b
    def multiply(self, a, b):
        return a * b
m = Math()
m.add(10,20)
m.multiply(10, 20)
```

```python
class Math:
    @staticmethod #이걸 사용하면 클래스 이름을 사용하여 함수 호출 가능
    def add(a, b):
        return a + b
    @staticmethod
    def multiply(a, b):
        return a * b
m = Math()
#static method 사용시 함수처럼 클래스명을 호출 가능
Math.add(10,20)
Math.multiply(10, 20)
```

 #### **Class Inheritance (상속)**

  - 기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스를 정의한다.

  - 상속 받고자 하는 대상인 기존 클래스는 (Parent, Super, Base class 라고 부른다.)

#### **method override**
 - 부모 클래스의 method를 재정의(override)
 - 하위 클래스(자식 클래스) 의 인스턴스로 호출시, 재정의된 메소드가 호출됨

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def eat(self, food):
        print(f'{self.name}은 {food}를 먹습니다.')
    def sleep(self, minute):
        print(f'{self.name}은 {minute}분 동안 잡니다.')
    def work(self, minute):
        print(f'{self.name}은 {minute}분 동안 일합니다.')
    
class Student(Person): #상속 받고자 하는 class를 적어줌 'Person'
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute):
        print(f'{self.name}은 {minute}분 동안 공부합니다.')
        
class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute):
        print(f'{self.name}은 {minute}분 동안 업무를 합니다.')

bob = Student('Bob', 25) 
bob.eat('BBQ') #Student 클래스에 eat이랑 sleep 함수가 없어도 상속받음
bob.sleep(30)
bob.work(60) #Student 클래스에서 메소드 오버라이드 
```

#### super 
 - 하위클래스(자식 클래스)에서 부모클래스의 method를 호출할 때 사용

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def eat(self, food):
        print(f'{self.name}은 {food}를 먹습니다.')
    def sleep(self, minute):
        print(f'{self.name}은 {minute}분 동안 잡니다.')
    def work(self, minute):
        print(f'{self.name}은 {minute}분 동안 일할 준비를 합니다.')
    
class Student(Person): #상속 받고자 하는 class를 적어줌 'Person'
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute):
        super().work(minute) #부모의 함수를 호출함
        print(f'{self.name}은 {minute}분 동안 공부합니다.')
        
class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def work(self, minute):
        super().work(minute) #부모의 함수를 호출함
        print(f'{self.name}은 {minute}분 동안 업무를 합니다.')

bob = Student('Bob', 25) 
bob.eat('BBQ') #Student 클래스에 eat이랑 sleep 함수가 없어도 상속받음
bob.sleep(30)
bob.work(60) #Student 클래스에서 메소드 오버라이드 

rob = Employee('Bob', 25) 
rob.eat('BBQ') #Student 클래스에 eat이랑 sleep 함수가 없어도 상속받음
rob.sleep(30)
rob.work(60) #Student 클래스에서 메소드 오버라이드 
```

```python
# Point
# 각 2차원 좌표평면 각 점(x, y)
# 연산
# 두 점의 덧셈, 뺄셈 (1, 2) + (3, 4) = (4, 6)
# 한점과 숫자의 곱셈 (1, 2) * 3 = (3, 6)
# 그 점의 길이 (0, 0) 부터의 거리
# x, y 값 가져오기
# 출력하기

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, pt): #언더바 __언더바__ 해주면 됨
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    def __sub__(self, pt): #언더바 __언더바__ 해주면 됨
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point(new_x, new_y)
    def get_x(self):
        return self.x
    def get_y(self):
        return self.y
    def multiply(self, factor):
        return Point(self.x * factor, self.y * factor)
    def __str__(self):
        return f'{self.x}, {self.y}'
    def length(self):
        return self.x ** 2 + self.y **2
              
p1 = Point(3,4)
p2 = Point(2,7)

print(p1) 
print(p2)
print(p1.multiply(3))
print(p1.length())
print(p1.get_x())
print(p1.get_y())
```

```python
# Point
# 각 2차원 좌표평면 각 점(x, y)
# 연산
# 두 점의 덧셈, 뺄셈 (1, 2) + (3, 4) = (4, 6)
# 한점과 숫자의 곱셈 (1, 2) * 3 = (3, 6)
# 그 점의 길이 (0, 0) 부터의 거리
# x, y 값 가져오기
# 출력하기

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, pt): #언더바 __언더바__ 해주면 됨
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    def __sub__(self, pt): #언더바 __언더바__ 해주면 됨
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point(new_x, new_y)
    def __mul__(self, factor):
        return Point(self.x * factor, self.y * factor)
    def __str__(self):
        return f'{self.x}, {self.y}'
    def __len__(self):
        return self.x ** 2 + self.y **2
    def __getitem__(self, index):
        if index == 0:
            return self.x
        if index == 1:
            return self.y
        else: 
            return False
              
p1 = Point(3,4)
p2 = Point(2,7)

print(p1) 
print(p2)
print(p1 * 3)
len(p1)
print(p1[0])
p1[1]
```

#### 연습문제)
 - 복소수 클래스를 정의 해봅시다.
 - 덧셈, 뺄셈, 곱셈 연산자 지원
 - 길이 (복소수의 크기) 지원 
 - 복소수 출력 '1 + 4j'와 같이 표현
 - 비교 연산 ==, != 지원
 - >=, <= , <, > 연산 지원
 - 절대값 지원


```python
import math
class ComplexNumber:
    def __init__(self, real, img): #실수부-real, 허수부-img
        self.real = real
        self.img = img
    def __add__(self, cn):
        return ComplexNumber(self.real + cn.real, self.img + cn.img)
    def __sub__(self, cn):
        return ComplexNumber(self.real - cn.real, self.img - cn.img)
    def __mul__(self, x):
        if type(x) == int:
            return ComplexNumber(self.real * x, self.img * x)
        elif type(x) == ComplexNumber:
            return ComplexNumber(self.real * x.real - self.img * x.img, self.real * x.img + self.img * x.real)
        #(a + bj)*(c + dj) = (ac -bd) + (ad + bc)j
    def __str__(self):
        if self.img > 0:
            return f'{self.real} + {self.img}j'
        else:
            return f'{self.real} - {abs(self.img)}j'
    def __eq__(self, cn):
        return self.real == cn.real and self.img == cn.img
    def __ne__(self, cn):
#         return self.real != cn.real or self.img != cn.img
        return not (self.real == cn.real and self.img == cn.img)
    def __abs__(self):
        return math.sqrt(self.real ** 2 + self.img ** 2)    
    def __len__(self):
        return (self.real ** 2 + self.img ** 2)

a = ComplexNumber(1,2)
b = ComplexNumber(3,5)
c = ComplexNumber(1,2)
print(a)
print(b)
print(a + b)
print(a - b)
print(a * b)
print(a * 3)
print(a == c)
print(a != b)
print(abs(a))
print(len(a))
```
