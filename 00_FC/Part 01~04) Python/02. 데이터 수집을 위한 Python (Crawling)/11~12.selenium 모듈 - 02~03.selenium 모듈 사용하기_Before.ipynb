```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
```

```python
driver = webdriver.Chrome(service=service)
driver.get('https://www.python.org')

time.sleep(2)

driver.close()
```

```python
driver = webdriver.Chrome(service=service)
driver.get('https://www.python.org')

search = driver.find_element('id', 'id-search-field')

search.clear()
time.sleep(3)
search.send_keys('lambda')
time.sleep(3)
search.send_keys(Keys.RETURN)
time.sleep(3)

driver.close()
```

```python
driver = webdriver.Chrome(service=service)
driver.get('https://n.news.naver.com/mnews/article/001/0014629543?rc=N&ntype=RANKING')

WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'span.u_cbox_count')))

comment = soup.find('span', class_='u_cbox_count')
comment.get_text()

driver.close()
```
