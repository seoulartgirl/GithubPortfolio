```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 그래프 데이터 생성

```python
x = np.linspace(0,10,11)
y = x**2 + x+2+np.random.randn(11)

print(x)
print(y)
```

### 그래프 출력하기
 - plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
   - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
   - 기본적으로, x, y에 해당하는 값이 필요

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

### 그래프에 주석 추가 

* **x, y 축 및 타이틀**

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('XY')
plt.plot(x,y)
```

* **grid 추가**

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.plot(x,y)
```

* **x, y축 범위 지정**

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)

plt.xlim(0,20)
plt.ylim(0,200)

plt.plot(x,y)
```

### subplot으로 여러 그래프 출력하기
 - subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(y,x,'k')
plt.subplot(2,2,4)
plt.plot(x,np.exp(x),'b')
```

```python
plt.subplot(2,3,1)
plt.plot(x,y,'r')

plt.subplot(2,3,2)
plt.plot(x,y,'g')

plt.subplot(2,3,3)
plt.plot(x,y,'r')

plt.subplot(2,3,4)
plt.plot(y,x,'k')

plt.subplot(2,3,5)
plt.plot(x, np.exp(x), 'b')

plt.subplot(2,3,6)
plt.plot(x,y,'g')

plt.show()
```

pi/4 : 1 = (4분원 안에 생성된 점의 개수): 전체 시도 횟수

pi/4 * (전체 시도 횟수) = (4분원 안에 생성된 점의 개수)

pi = 4 * (4분원 안에 생성된 점의 개수) / (전체 시도 횟수)

pi = 4 * (4분원 안에 생성된 점의 개수) / 1e7 # 1e7(천만번)

```python
total = int(1e7)
total
```

```python
total = int(1e7)
points = np.random.rand(total, 2)
points.shape
```

```python
points # [x,y]를 명시
```

```python
#원 안에 들어가려면 x,y 제곱의 합이 1보다 작으면 원안에 들어감
np.sum(points ** 2, axis=1)
```

```python
np.sum(points ** 2, axis=1) < 1
```

```python
# True의 개수를 구해줌
np.sum(np.sum(points ** 2, axis=1) < 1)
```

```python
#파이 근사치와 유사 (3.141592)
4 * np.sum(np.sum(points ** 2, axis=1) < 1) / total
```
