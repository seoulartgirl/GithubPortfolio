```python
# 데이터 불러오기
boston = pd.read_csv('data/Boston_house.csv')
boston.head()
```

```python
# target 제외한 데이터만 뽑기
boston_data = boston.drop(['Target'], axis=1)
```

```python
# data 통계 뽑아보기 
boston_data.describe()
```

```python
'''
타겟 데이터
1978 보스턴 주택 가격
506개 타운의 주택 가격 중앙값 (단위 1,000 달러)

특징 데이터
CRIM: 범죄율 *
INDUS: 비소매상업지역 면적 비율
NOX: 일산화질소 농도
RM: 주택당 방 수 *
LSTAT: 인구 중 하위 계층 비율 *
B: 인구 중 흑인 비율
PTRATIO: 학생/교사 비율
ZN: 25,000 평방피트를 초과 거주지역 비율
CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
AGE: 1940년 이전에 건축된 주택의 비율
RAD: 방사형 고속도로까지의 거리
DIS: 직업센터의 거리
TAX: 재산세율'''
```

# crim/rm/lstat 세게의 변수로 각각 단순 선형 회귀 분석하기

```python
## 변수 설정 target/crim/rm/lstat
target = boston[['Target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston[['LSTAT']]
```

## target ~ crim 선형회귀분석

```python
# crim변수에 상수항추가하기 
crim1 = sm.add_constant(crim, has_constant='add')
crim1.head()
```

```python
# sm.OLS 적합시키기 
model1 = sm.OLS(target, crim1)
fitted_model = model1.fit()
```

```python
# summary함수통해 결과출력
fitted_model.summary()
```

=> R-squared: 0.151 (범죄율은 15% 정도 비중)

=> CRIM: -0.4152 (범죄율에 해당하는 회귀 계수값은 범죄율이 1단위 증가할 때 y가 -0.41단위 증가함/0.41단위 감소함)

```python
## 회귀 계수 출력
fitted_model.params
```

### y_hat=beta0 + beta1 * X 계산해보기

```python
#회귀 계수 x 데이터(X)
np.dot(crim1, fitted_model.params) #y_hat 구하기
```

```python
## predict함수를 통해 yhat구하기
pred1 = fitted_model.predict(crim1)
```

```python
## 직접구한 yhat과 predict함수를 통해 구한 yhat차이 
np.dot(crim1, fitted_model.params) - pred1
```

## 적합시킨 직선 시각화 

```python
import matplotlib.pyplot as plt

plt.yticks(fontname = "Arial") #
plt.scatter(crim,target,label="data")
plt.plot(crim,pred1,label="result")
plt.legend()
plt.show()
```

```python
plt.scatter(target,pred1)
plt.xlabel("real_value")
plt.ylabel("pred_value")
plt.show()
```

```python
## residual 시각화 

fitted_model.resid.plot()
plt.xlabel("residual_number")
plt.show()
```

```python
##잔차의 합계산해보기
np.sum(fitted_model.resid) #거의 0에 수렴함
```

## 위와 동일하게 rm변수와 lstat 변수로 각각 단순선형회귀분석 적합시켜보기 

```python
# 상수항추가
rm1 = sm.add_constant(rm, has_constant='add')
lstat1 = sm.add_constant(lstat, has_constant='add')
```

```python
# 회귀모델 적합
model2 = sm.OLS(target, rm1)
fitted_model2 = model2.fit()
model3 = sm.OLS(target, lstat1)
fitted_model3 = model3.fit()
```

```python
# rm모델 결과 출력
fitted_model2.summary()
```

=> 주택당 방 수 48.4% 정도 비중

```python
# lstat모델 결과 출력 
fitted_model3.summary()
```

=> 인구 중 하위 계층 비율 '54.4%'

```python
## 각각 yhat_예측하기 
pred2 = fitted_model2.predict(rm1)
pred3 = fitted_model3.predict(lstat1)
```

```python
## rm모델 시각화
import matplotlib.pyplot as plt
plt.scatter(rm,target,label="data")
plt.plot(rm,pred2,label="result")
plt.legend()
plt.show()
```

```python
## lstat모델 직선 시각화
import matplotlib.pyplot as plt
plt.scatter(lstat,target,label="data")
plt.plot(lstat,pred3,label="result")
plt.legend()
plt.show()
```

```python
# rm모델 reisidual 시각화 
fitted_model2.resid.plot()
plt.xlabel("residual_number")
plt.show()
```

```python
# lstat모델 residual시각화
fitted_model3.resid.plot()
plt.xlabel("residual_number")
plt.show()
```
