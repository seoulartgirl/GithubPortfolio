```python
items = soup.findAll('item')
print(len(items))
items[0]
```

```python
festival = []

for item in soup.findAll('item'):
    try:
        title = item.find('title').text
        addr = item.find('addr1').text
        tel = item.find('tel').text
        festival.append([title, addr, tel])
    except:
        print('No data')
    
festival
```

- 조회 정보를 데이터프레임으로 저장

```python
df = pd.DataFrame(festival, columns=['축제이름','주소','연락처'])
df
```
