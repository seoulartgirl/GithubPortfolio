### 연습문제1. 
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과 후반기(7월-12월)실적을 나타내는 데이터 프레임을 작성한 후 결합한다.


- 실적 정보는 "매출","비용","이익"으로 이루어진다.

- 이익=매출-비용 : 가공필드로 생성
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

#### 1) month 변수 생성
- 리스트 컴프리헨션을 사용

```python
month = [str(i)+'월' for i in range(1,13)]
month
```

```python
#결과값 참고
```

#### 2) 상반기(1~6월) 매출과 비용 data를 정수 난수를 이용하여 생성

- 1~6월의 매출과 비용은 딕셔너리로 구성하되, 키값은 month변수를 활용
- 매월의 매출과 비용은 numpy의 1~100사이의 정수 난수로 생성

```python
# 매출, 비용 data random  생성
data = {f'{month}월': np.random.randint(1, 100, 2) for month in range(1, 7)}
data
```

```python
# 매출, 비용 data random  생성
data = {}
for i in month[0:6]:
    data[i] = np.random.randint(1, 100, 2)
data
```

```python
# 결과값 참고
# 매출, 비용 data random  생성
```

#### 3) 상반기(1~6월)의 매출과 비용 데이터프레임 생성

- 2)번에서 생성한 딕셔너리 데이터로 데이터프레임 생성

```python
# 데이터프레임 생성

df = pd.DataFrame(data, index=['매출','비용'])
df
```

```python
# 결과값 참고
# 데이터프레임 생성
```

#### 4) '이익' 행 추가
- '이익' 행이 추가된 데이터 프레임 작성
- 이익 = 매출 - 비용 

```python
# '이익' 행 생성
df.loc['이익'] = df.loc['매출'] - df.loc['비용']
df
```

```python
# 결과값 참고
# '이익' 행 생성
```

#### 5) 하반기(7~12월) 매출과 비용 data를 정수 난수를 이용하여 생성

- 7~!2월의 매출과 비용은 딕셔너리로 구성하되, 키값은 month변수를 활용
- 매월의 매출과 비용은 numpy의 1~100사이의 정수 난수로 생성

```python
# 매출, 비용 data random  생성
data2 = {f'{month}월': np.random.randint(1,100,2) for month in range(7,13)}
data2
```

```python
data2={}
for i in month[6:]:
    data2[i] = np.random.randint(1, 100, 2)
data2
```

```python
#결과값 참고
# 매출, 비용 data random  생성
```

#### 6) 하반기(7~12월) 실적 데이터프레임 생성
- 매출, 비용으로 구성

```python
# 데이터프레임 생성
df2 = pd.DataFrame(data2, index=['매출', '비용'])
df2
```

```python
# 결과값 참고
# 데이터프레임 생성
```

#### 7) 하반기 실적에 '이익' 행 추가
- '이익' 행이 추가된 데이터 프레임 작성
- 이익 = 매출 - 비용 

```python
# 가공 행 이익 생성
df2.loc['이익'] = df2.loc['매출'] - df2.loc['비용']
df2
```

```python
# 결과값 참고
# 가공 행 이익 생성
```

#### 8) 상반기 실적 데이터프레임과 하반기 실적 데이터프레임 연결

```python
df3 = pd.concat([df, df2], axis=1)
#또는 df1.join(df2) 
#또는 df1.merge(df2, left_index=True, right_index=True)
df3
```

```python
#결과값 참고
```

#### 9) '총실적' 열 추가

```python
df3['총실적'] = df3[['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']].sum(axis=1).astype(int)
df3
```

```python
df3['총실적'] = df3.sum(axis=1).astype(int)
df3
```
