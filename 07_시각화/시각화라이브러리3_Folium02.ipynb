## 5. 지도에 마커(marker) 설정 

- POI(Point Of Interest, 관심지점)을 추가하고 관련 정보 제공
- 마커(marker) : 특정 위치를 표시하는 표식

### 마커를 부착한 지도 생성

단계1. 마커 생성   
- **folium.Marker(location=[위도,경도], popup='', tooltip='', icon=아이콘객체)**
   - popup : 마커 클릭 시 표시되는 문구
       - 환경에 따라 지원 안되는 경우도 있음
   - tooltip : 마우스 hover시 표기되는 문구
   - icon : Marker 형태 결정
       - 기본값 : information icon
                
단계2. 마커 아이콘 생성
- **folium.Icon(color='색상', icon='아이콘 모양')**

    
   
단계3. 마커 생성 후 부착
- **folium.Marker().add_to(지도객체변수)**

#### Map 객체 생성

```python
map1 = folium.Map(location=[37.56, 127.05], zoom_start=10)
map1
```

#### Marker 객체 생성 및 Map에 부착

```python
folium.Marker((37.547433727420454, 127.04803964585416),
             popup='Subway', tooltip='뚝섬역', 
              icon=folium.Icon(icon='cloud')).add_to(map1)
map1
```

```python
#뚝섬역: 37.547433727420454, 127.04803964585416
#과천 서울대공원: 37.42769504509449, 127.01700374788183
#카카오 판교오피스: 37.39598490029507, 127.11053956708399
#멀티캠퍼스 세종캠: 37.54817134388939, 127.0733980772467

locs= {'뚝섬역' : (37.547433727420454, 127.04803964585416), 
        '과천서울대공원': (37.42769504509449, 127.01700374788183),
        '카카오판교': (37.39598490029507, 127.11053956708399),
        '멀티캠퍼스 세종': (37.54817134388939, 127.0733980772467)}
```

```python
for name, loc in locs.items():
    folium.Marker(loc,
             popup=name, tooltip=name, 
              icon=folium.Icon(icon='cloud')).add_to(map1)
map1
```

### 마커 클릭 시 특정 사이트로 이동하기

```python
lat, lon = [37.40, 127.09]
zoom_size = 13
map2 = folium.Map([lat,lon], zoom_start=zoom_size)
folium.TileLayer(tiles='Stamen Terrain').add_to(map2)
folium.Marker(location=(37.42769504509449, 127.01700374788183),
             popup = '<a href="https://grandpark.seoul.go.kr/main/ko.do" target=_blank>과천서울대공원</a>',
              tooltip='서울대공원', 
              icon=folium.Icon(icon='info-sign')).add_to(map2)

folium.Marker(location=(37.54817134388939, 127.0733980772467),
             popup = '<a href="https://www.multicampus.com/publish/popup/multicampus_map.html" target=_blank>멀티캠세종</a>',
              tooltip='멀티캠퍼스 세종', 
              icon=folium.Icon(icon='info-sign')).add_to(map2)
map2
```

### 마커에 custom icon 적용하기 

- icon옵션 custom
    - prefix 옵션 : 두 가지 - fa, glyphicon
    - prefix='fa' 옵션을 선택할 때 icon 선택하는 사이트
        - https://fontawesome.com/
    - prefix = 'glyphicon' 옵션을 선택할 때 icon 선택하는 사이트
        - http://bootstrapk.com/components/

```python
lat, lon = [37.40, 127.09]
zoom_size = 9

#지도 객체 생성
map3 = folium.Map([lat,lon], zoom_start=zoom_size)

#지도 타일 지정
folium.TileLayer(tiles='Stamen Terrain').add_to(map3)

#Marker를 지정하기 위한 Icon 객체 생성
icon = folium.Icon(color='black', icon='info-sign')

#Marker 객체 생성 후 Map 객체에 추가
folium.Marker(location=(37.42769504509449, 127.01700374788183),
             popup = '<a href="https://grandpark.seoul.go.kr/main/ko.do" target=_blank>과천서울대공원</a>',
              tooltip='서울대공원', 
              icon=icon).add_to(map3)

#prefix='fa'
icon = folium.Icon(color='green', icon='building')
folium.Marker(location=(37.54817134388939, 127.0733980772467),
             popup = '<a href="https://www.multicampus.com/publish/popup/multicampus_map.html" target=_blank>멀티캠세종</a>',
              tooltip='멀티캠퍼스 세종', 
              icon=icon).add_to(map3)

icon = folium.Icon(color='purple', icon='info-sign')
folium.Marker(location=(37.39598490029507, 127.11053956708399),
             popup = '<a href="https://www.kakaocorp.com/page/" target=_blank>카카오</a>',
              tooltip='카카오', 
              icon=icon).add_to(map3)
map3
```

## 6. 지도에 원 그리기

: 지도에 데이터 값, 영역 등의 정보를 표현하기 위한 방법

- folium.Circle()
- folium.CircleMarker()

#### Circle()  생성

```python
lat, lon = 37.54, 127.05469
zoom_size = 15
m = folium.Map(location=[lat,lon], zoom_start=zoom_size)
m
```

```python
#circle
folium.Circle(location=[lat, lon], radius=100, color='red', 
              fill_color='blue', popup='Circle popup', 
             tooltip='Circle tooltip').add_to(m)
m
```

#### CircleMarker 생성

```python
lat, lon = 37.54, 127.05469
zoom_size = 15
m = folium.Map(location=[lat,lon], zoom_start=zoom_size)

#circle
folium.CircleMarker(location=[lat, lon], radius=50, 
                    color='#34ebc6', 
                    fill_color='#c634eb', 
                    popup='Circle popup', 
                     tooltip='Circle tooltip').add_to(m)
m
```

- MarkerCluster 추가

```python
from folium.plugins import MarkerCluster
```

```python
#Map 객체 생성
lat, lon = 37.566535, 126.9779691999996
m4 = folium.Map([lat,lon], zoom_start=16)

tiles = 'https://tiles.stadiamaps.com/tiles/osm_bright/{z}/{x}/{y}{r}.png'
attr = 'Stadia Maps'
folium.TileLayer(tiles=tiles, attr=attr).add_to(m4)

#MarkerCluster 객체 생성하여 지도에 붙이기
marker_cluster = MarkerCluster().add_to(m4)

#Marker를 MarkerCluster에 붙이기
for name, loc in gu_loc.items():
    folium.Marker(location=loc,
             popup=name, tooltip=name, 
              icon=folium.Icon(color='red', 
                icon='info-sign')).add_to(marker_cluster)
m4
```
